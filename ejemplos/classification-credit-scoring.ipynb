# Curva ROC del modelo

```python
model2.predict(X_test)
```

```python
model2.predict_proba(X_test)
```

```python
#TP True Positives
#TN True Negatives
#P Total Positives
#N Total Negatives

clase_positiva = 1
P = np.sum(y_test==clase_positiva)
N = np.sum(y_test!=clase_positiva)
P, N

ps = model2.predict_proba(X_test)[:, clase_positiva]
ps
```

```python
umbral = 0.05  # No le dohy eel credito si p(impago) >= 0.1
sum(ps < umbral)
```

```python
sum(y_test[ps < umbral]==0)
```

```python
umbrales = np.unique(ps)

TPRs, FPRs = [], []
for u in umbrales:
    alarma = (ps >= u) # casos en los que doy alarma (no presto)

    TP = sum(alarma & (y_test==clase_positiva))
    FP = sum(alarma & (y_test!=clase_positiva))
    TPR = TP / P # True Positive Rate
    FPR = FP / N # False Positive Rate
    
    TPRs.append(TPR)
    FPRs.append(FPR)
```

```python

plt.figure(figsize=(7,10))
plt.subplot(2,1,1)

plt.plot(FPRs, TPRs, ".")
plt.title("Curva ROC")
plt.xlabel("FPR")
plt.ylabel("TPR")


plt.subplot(2,1,2)
plt.plot(FPRs, umbrales, "-r")
plt.title("Umbrales")
plt.xlabel("FPR")
plt.ylabel("Umbral")

plt.show()
```
